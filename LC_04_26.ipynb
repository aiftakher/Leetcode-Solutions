```python
# 204

class Solution:
    def countPrimes(self, n: int) -> int:
        if n < 3:
            return 0
        primes = [True] * (n)
        primes[0] = primes[1] = False
        for i in range(2, int(n ** 0.5) + 1):
            #print(i)
            if primes[i]:
                start = i * 2
                while start < n:
                    primes[start] = False
                    start += i
        return sum(primes)

sol = Solution()
print(sol.countPrimes(10)) # 4
```

```python
# LC 509
class Solution:
    def fib(self, n: int) -> int:
        def recurse(n):
            if n==0:
                return dp[0]
            if n==1:
                return dp[1]
            if dp[n] > 0:
                return dp[n]
            dp[n] = recurse(n-1) + recurse(n-2)
            return dp[n]
        
        dp = [0]*(n+1)
        dp[0] = 0
        if n>=1: 
            dp[1] = 1

        return recurse(n) 
    
sol = Solution()
print(sol.fib(0))
```

```python
# LC 258

class Solution:
    def addDigits(self, num):
        s = str(num)
#        print(s)
        while len(s) > 1:
            sum = 0
            for i in range(len(s)):
                sum += int(s[i])
 #           print(sum)
            s = ""
            s = str(sum)
        return int(s)

sol = Solution()
num = 38
print(sol.addDigits(num))
        
```

```python
# LC 873

class Solution:
    def lenLongestFibSubseq(self, arr):

        s = set(arr)
        n = len(arr)
        res = 0
        for i in range(n):
            for j in range(i+1, n):
                a = arr[i]
                b = arr[j]
                c = a + b
                l = 2
                while c in s:
                    a = b
                    b = c
                    c = a + b
                    l += 1
                res = max(res, l)
        return res if res > 2 else 0



sol = Solution()
arr = [1,2,3,4]
print(sol.lenLongestFibSubseq(arr))
```
